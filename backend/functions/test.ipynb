```python
testid = 'spotify:track:4zrKN5Sv8JS5mqnbVcsul7'

def get_track_data(trackid):
    try:
        data = sp.track(trackid)
        name = data['name']
        artist = data['artists'][0]['name']
        album_art = data['album']['images'][0]['url']
        return name, artist, album_art
    except:
        return

get_track_data(testid)
```

```python
def get_currently_playing():
    data = sp.current_user_playing_track()
    if data:
        name, artist, album_art = get_track_data(data['item']['id'])
        progress_ms = data['progress_ms']/1000
        duration = data['item']['duration_ms']/1000
        return name, artist, album_art, progress_ms, duration
    return

    
get_currently_playing()
```

```python
def get_search_results(query, limit=50):
    tracks = sp.search(q=query, type='track',limit=limit)['tracks']['items']
    results = []
    for track in tracks:
        results.append((track['id'], track['name'], track['artists'][0]['name'], track['album']['images'][0]['url']))
    return results


results = get_search_results('celestial')
results
```
